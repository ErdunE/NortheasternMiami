# **Image Classification**

**Goal:** Build a DL-based for Digit Classification on Street View House Numbers dataset.  

**Street View House Numbers (SVHN**) is a digit classification benchmark dataset that contains 600,000 32Ã—32 RGB images of printed digits (from 0 to 9) cropped from pictures of house number plates. The cropped images are centered in the digit of interest, but nearby digits and other distractors are kept in the image. SVHN has three sets: training, testing sets and an extra set with 530,000 images that are less difficult and can be used for helping with the training process.

**Question 1:** Load the images from file. Present the number of training, testing and Validation samples. Also, show the Image Dimensions, labels, and image formats.

```python
import os
import h5py
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from sklearn.model_selection import train_test_split

# Dataset paths
BASE_DIR = os.path.join(os.getcwd(), "Assignment9Dataset")
TRAIN_DIR = os.path.join(BASE_DIR, "train", "train")
TEST_DIR  = os.path.join(BASE_DIR, "test", "test")
EXTRA_DIR = os.path.join(BASE_DIR, "extra", "extra")

TRAIN_MAT = os.path.join(BASE_DIR, "train_digitStruct.mat")
TEST_MAT  = os.path.join(BASE_DIR, "test_digitStruct.mat")
EXTRA_MAT = os.path.join(BASE_DIR, "extra_digitStruct.mat")

for p in [BASE_DIR, TRAIN_DIR, TEST_DIR, EXTRA_DIR, TRAIN_MAT, TEST_MAT]:
    print("Path exists:", p, os.path.exists(p))

# Helper to list image files
def list_images(folder):
    if not os.path.isdir(folder):
        return []
    return sorted([f for f in os.listdir(folder) if f.lower().endswith(('.png', '.jpg', '.jpeg'))])

train_files = list_images(TRAIN_DIR)
test_files  = list_images(TEST_DIR)
extra_files = list_images(EXTRA_DIR)

print("\nDataset File Counts:")
print(f"  Train : {len(train_files)}")
print(f"  Test  : {len(test_files)}")
print(f"  Extra : {len(extra_files)}")


# Read labels from .mat (HDF5 format)
def _read_ascii_string_from_ref(h5f, ref):
    arr = np.array(h5f[ref][()]).squeeze()
    return ''.join(chr(int(c)) for c in arr)

def _read_numeric_vector_from_ref(h5f, ref):
    data = np.array(h5f[ref][()]).squeeze()
    if data.ndim == 0:
        return [int(data)]
    return [int(x) for x in data.ravel()]

def _coerce_to_int_list(h5f, obj):
    if hasattr(obj, 'dtype'):
        val = obj[()]
    else:
        val = obj
    val = np.array(val)
    if val.dtype == np.object_:
        out = []
        for r in val.ravel():
            out.extend(_read_numeric_vector_from_ref(h5f, r))
        return out
    if np.issubdtype(val.dtype, np.number):
        val = val.squeeze()
        if val.ndim == 0:
            return [int(val)]
        return [int(x) for x in val.ravel()]
    if val.shape == ():
        elem = val.item()
        if isinstance(elem, h5py.Reference):
            return _read_numeric_vector_from_ref(h5f, elem)
        if isinstance(elem, (int, float, np.integer, np.floating)):
            return [int(elem)]
    raise TypeError(f"Unsupported HDF5 content: dtype={val.dtype}, shape={val.shape}")

def read_digitstruct_labels(h5path):
    labels_by_name = {}
    with h5py.File(h5path, 'r') as f:
        bbox_ds = f['/digitStruct/bbox']
        name_ds = f['/digitStruct/name']
        for i in range(name_ds.shape[0]):
            name_ref = name_ds[i][0]
            filename = _read_ascii_string_from_ref(f, name_ref)
            bbox_ref = bbox_ds[i][0]
            bbox_obj = f[bbox_ref]
            raw_labels = _coerce_to_int_list(f, bbox_obj['label'])
            labels = [0 if x == 10 else x for x in raw_labels]
            labels_by_name[filename] = labels
    return labels_by_name

train_labels_map = read_digitstruct_labels(TRAIN_MAT)
test_labels_map  = read_digitstruct_labels(TEST_MAT)

# Create Validation Split (80/20)
train_files_split, val_files_split = train_test_split(
    train_files, test_size=0.2, random_state=42
)

print("\nDataset Split Summary:")
print(f"  Training samples   : {len(train_files_split)}")
print(f"  Validation samples : {len(val_files_split)}")
print(f"  Testing samples    : {len(test_files)}")

# Inspect one image
def describe_image_info(image_path):
    img = Image.open(image_path)
    arr = np.array(img)
    h, w = img.height, img.width
    channels = arr.shape[2] if arr.ndim == 3 else 1
    mode = img.mode
    return (w, h, channels, mode)

if len(train_files_split) > 0:
    sample_path = os.path.join(TRAIN_DIR, train_files_split[0])
    w, h, c, mode = describe_image_info(sample_path)
    print("\nExample image info:")
    print(f"  Filename       : {train_files_split[0]}")
    print(f"  Image size     : {w} x {h} (width x height)")
    print(f"  Num of channels: {c}")
    print(f"  Image mode     : {mode}")

# Visualize few samples with labels
plt.figure(figsize=(14, 4))
for i, fname in enumerate(train_files_split[:10]):
    img_path = os.path.join(TRAIN_DIR, fname)
    img = Image.open(img_path)
    labels = train_labels_map.get(fname, None)
    title = f"{fname}\nlabel(s): {labels}" if labels is not None else f"{fname}\nlabel(s): N/A"
    plt.subplot(2, 5, i + 1)
    plt.imshow(img)
    plt.axis("off")
    plt.title(title, fontsize=8)
plt.suptitle("Sample SVHN Training Images with Labels", fontsize=12)
plt.tight_layout()
plt.show()
```

The SVHN dataset was successfully loaded, including 33,402 training, 13,068 testing, and 20% (6,681) validation samples derived from the training set.
Each image is in RGB format, and sizes vary before preprocessing (e.g., 112Ã—41, 74Ã—135). The labels come from the HDF5 structure /digitStruct/name and /digitStruct/bbox, where the value 10 represents the digit â€œ0.â€

Images often contain multiple digits, so taking only the first label simplifies the problem to single-digit classification.
A separate validation set is created to tune hyperparameters and monitor overfitting, while the test set is reserved for final evaluation.
This datasetâ€™s variation in lighting, background, and number positions provides a realistic and challenging environment for CNN-based models.

**Question 2:** Using Keras or PyTorch, clean, normalize, and re-shape the data for the model input.

```python
import os
import numpy as np
import torch
import torch.nn.functional as F
from torchvision import transforms
from PIL import Image
import matplotlib.pyplot as plt

# Define transforms
transform = transforms.Compose([
    transforms.Resize((32, 32)),
    transforms.ToTensor(),              # scales to [0,1]
    transforms.Normalize((0.5, 0.5, 0.5),
                         (0.5, 0.5, 0.5))  # â†’ [-1,1]
])

# Load subset for demonstration
def load_images_and_labels(image_dir, labels_map, limit=200):
    images, labels = [], []
    files = sorted(os.listdir(image_dir))[:limit]
    for f in files:
        path = os.path.join(image_dir, f)
        if not os.path.isfile(path):
            continue
        img = Image.open(path).convert("RGB")
        img_t = transform(img)
        images.append(img_t)

        # Take the first digit only
        label_list = labels_map.get(f, [])
        if len(label_list) > 0:
            labels.append(label_list[0])

    X = torch.stack(images)
    y = torch.tensor(labels)
    return X, y

X_train, y_train = load_images_and_labels(TRAIN_DIR, train_labels_map, limit=500)
print("Preprocessed training subset shape:", X_train.shape)
print("Label tensor shape:", y_train.shape)
print("Pixel range: min =", X_train.min().item(), "max =", X_train.max().item())

# One-hot encode labels
num_classes = 10
y_train_onehot = F.one_hot(y_train, num_classes=num_classes).float()
print("Example one-hot label:", y_train_onehot[0])
print("One-hot shape:", y_train_onehot.shape)

# Visualization
fig, axes = plt.subplots(1, 2, figsize=(6,3))
# raw image
raw_img = Image.open(os.path.join(TRAIN_DIR, sorted(os.listdir(TRAIN_DIR))[0])).convert("RGB")
axes[0].imshow(raw_img)
axes[0].set_title("Original")
axes[0].axis("off")

# preprocessed
proc_img = (X_train[0].permute(1,2,0).numpy() * 0.5 + 0.5)  # inverse normalization
axes[1].imshow(proc_img)
axes[1].set_title("After Preprocessing")
axes[1].axis("off")
plt.show()

# show few preprocessed examples
plt.figure(figsize=(12,4))
for i in range(10):
    img = (X_train[i].permute(1,2,0).numpy() * 0.5 + 0.5)
    plt.subplot(2,5,i+1)
    plt.imshow(img)
    plt.axis("off")
    plt.title(f"Label: {y_train[i].item()}")
plt.suptitle("Sample Preprocessed (32Ã—32, normalized) Images", fontsize=12)
plt.tight_layout()
plt.show()
```

All SVHN images were resized to 32Ã—32Ã—3 to match CNN input requirements.
torchvision.transforms.ToTensor() converted images from [0,255] RGB values to the [0,1] floating-point range and reordered dimensions to [C, H, W].
Then, an additional normalization step (x - 0.5) / 0.5 mapped all values into [-1, 1], stabilizing gradients and improving convergence under ReLU and BatchNorm layers.

Each label was represented as a 10-dimensional one-hot vector, compatible with the softmax output layer for digit classification.
Visual inspection confirms that digits remain clear after resizing and normalization, ensuring consistent inputs across samples.

Overall, this preprocessing pipeline provides uniform dimensions, numerically stable input ranges, and proper label encoding, forming a clean foundation for CNN training.


**Question 3:** Data augmentation in deep learning is a technique to artificially expand a dataset by creating modified copies of existing data, thereby increasing the size and diversity of the training set. This process improves model performance by enhancing generalizability and reducing overfitting, as it allows the model to encounter a wider range of real-world variations. Common methods include geometric transformations (like image rotation and flips), color space transformations (adjusting brightness, contrast), and random erasing.

Apply any Data Augmentation to increase the variety of training examples, lead to help the model generalize better. Provide the description of apparoch and the importance of data augmentation in DL process.


```python
import torch
from torchvision import transforms
from PIL import Image
import matplotlib.pyplot as plt
import random, os

# Define three augmentation levels
# No augmentation (baseline)
no_aug = transforms.Compose([
    transforms.Resize((32, 32)),
    transforms.ToTensor(),
    transforms.Normalize((0.5,)*3, (0.5,)*3)
])

# Light augmentation
light_aug = transforms.Compose([
    transforms.Resize((32, 32)),
    transforms.RandomRotation(10),
    transforms.ColorJitter(brightness=0.2, contrast=0.2),
    transforms.ToTensor(),
    transforms.Normalize((0.5,)*3, (0.5,)*3)
])

# Heavy augmentation
heavy_aug = transforms.Compose([
    transforms.Resize((32, 32)),
    transforms.RandomRotation(45),
    transforms.RandomAffine(degrees=0, translate=(0.2, 0.2), scale=(0.8, 1.2)),
    transforms.ColorJitter(brightness=0.5, contrast=0.5, saturation=0.4),
    transforms.RandomHorizontalFlip(0.5),
    transforms.ToTensor(),
    transforms.Normalize((0.5,)*3, (0.5,)*3)
])

# Visualize augmentation on a single image
def preview_augmentation(image_path, transforms_list, titles):
    """Visualize one image under different augmentation settings."""
    original_img = Image.open(image_path).convert("RGB")

    plt.figure(figsize=(12, 3))
    for i, (transform, title) in enumerate(zip(transforms_list, titles)):
        aug_img_tensor = transform(original_img)
        aug_img = (aug_img_tensor.permute(1,2,0).numpy() * 0.5 + 0.5).clip(0,1)
        plt.subplot(1, 3, i + 1)
        plt.imshow(aug_img)
        plt.axis("off")
        plt.title(title)
    plt.suptitle("Augmentation Strength Comparison", fontsize=13)
    plt.tight_layout()
    plt.show()

# Randomly choose one image for demonstration
sample_fname = random.choice(os.listdir(TRAIN_DIR))
sample_path = os.path.join(TRAIN_DIR, sample_fname)

preview_augmentation(
    sample_path,
    [no_aug, light_aug, heavy_aug],
    ["No Augmentation", "Light Augmentation", "Heavy Augmentation"]
)

# Show multiple examples per augmentation level
def show_random_augments(image_dir, transform, title, n=5):
    """Display multiple augmented samples."""
    files = sorted(os.listdir(image_dir))[:n]
    plt.figure(figsize=(12, 3))
    for i, fname in enumerate(files):
        path = os.path.join(image_dir, fname)
        img = Image.open(path).convert("RGB")
        img_aug = transform(img)
        img_aug = (img_aug.permute(1,2,0).numpy() * 0.5 + 0.5).clip(0,1)
        plt.subplot(1, n, i + 1)
        plt.imshow(img_aug)
        plt.axis("off")
        plt.title(fname)
    plt.suptitle(title, fontsize=12)
    plt.show()

show_random_augments(TRAIN_DIR, light_aug, "Light Augmentation Examples")
show_random_augments(TRAIN_DIR, heavy_aug, "Heavy Augmentation Examples")
```

Three augmentation levels were implemented using `torchvision.transforms`:

1. **No Augmentation** â€“ baseline with only resize and normalization.  
2. **Light Augmentation** â€“ Â±10Â° rotation and mild brightness/contrast jitter.  
3. **Heavy Augmentation** â€“ Â±45Â° rotation, translation (20%), scaling (0.8â€“1.2Ã—), strong color jitter, and horizontal flips.

Light augmentation produces realistic variations similar to natural photographic noiseâ€”such as small tilts and lighting differencesâ€”while maintaining label integrity.  
Heavy augmentation introduces more radical transformations, increasing robustness to viewpoint and brightness shifts but at the cost of occasional semantic distortion.

Visual inspection confirms that all augmented samples preserve the digitsâ€™ structure while diversifying the dataset.  
This process improves model generalization by teaching the CNN to recognize digits invariant to small rotations, translations, and illumination changes, thus reducing overfitting and enhancing robust feature learning.

**Question 4:** Normalize, reshape, and encode the data to make it ready for classification.
Also, what is data batching and why we need it? perform it.

```python
import os
import torch
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
from PIL import Image
import matplotlib.pyplot as plt

# Define Custom Dataset for SVHN
class SVHNDataset(Dataset):
    def __init__(self, image_dir, labels_map, transform=None, limit=None):
        self.image_dir = image_dir
        self.labels_map = labels_map
        self.transform = transform
        # Collect image files
        files = sorted([
            f for f in os.listdir(image_dir)
            if f.lower().endswith(('.png', '.jpg', '.jpeg'))
        ])
        if limit:
            files = files[:limit]
        self.files = files

    def __len__(self):
        return len(self.files)

    def __getitem__(self, idx):
        fname = self.files[idx]
        fpath = os.path.join(self.image_dir, fname)
        img = Image.open(fpath).convert("RGB")
        label_list = self.labels_map.get(fname, [0])
        label = label_list[0]  # pick the first digit
        if self.transform:
            img = self.transform(img)
        return img, label

# Define normalization and transformation pipeline
transform = transforms.Compose([
    transforms.Resize((32, 32)),                
    transforms.ToTensor(),                      
    transforms.Normalize((0.5,)*3, (0.5,)*3)    
])

# Create datasets and dataloaders
BATCH_SIZE = 64

train_dataset = SVHNDataset(TRAIN_DIR, train_labels_map, transform=transform, limit=2000)
test_dataset  = SVHNDataset(TEST_DIR,  train_labels_map, transform=transform, limit=500)

train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
test_loader  = DataLoader(test_dataset,  batch_size=BATCH_SIZE, shuffle=False)

print(f"Train: {len(train_dataset)} samples | Batch size = {BATCH_SIZE} | Total batches = {len(train_loader)}")
print(f"Test : {len(test_dataset)} samples  | Total batches = {len(test_loader)}")

# Visualize one batch to confirm normalization & batching
def show_batch(images, labels, n=8):
    imgs = images[:n].permute(0, 2, 3, 1)   
    imgs = imgs * 0.5 + 0.5                 
    plt.figure(figsize=(12, 3))
    for i in range(n):
        plt.subplot(1, n, i + 1)
        plt.imshow(imgs[i])
        plt.title(f"Label: {labels[i].item()}")
        plt.axis("off")
    plt.suptitle("Example Batch from DataLoader", fontsize=12)
    plt.show()

# display a sample batch
sample_imgs, sample_labels = next(iter(train_loader))
show_batch(sample_imgs, sample_labels)
```

A custom `SVHNDataset` class and PyTorch `DataLoader` were implemented to prepare the SVHN dataset for model training.  
Each image was resized to 32Ã—32 pixels and normalized to the range [âˆ’1, 1] using  
`transforms.Normalize((0.5,)*3, (0.5,)*3)`, ensuring consistent scale and stable gradients.

The `DataLoader` divides the dataset into mini-batches (batch size = 64) and shuffles them each epoch.  
Batching improves training efficiency by processing multiple samples in parallel and enhances gradient stability through averaging over a batch.  
This leads to smoother optimization and prevents overfitting to individual samples.

The visualization of one batch confirms correct normalization, reshaping, and label alignment across samples, ensuring the dataset is ready for CNN model training.

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
import time

# Define Simple CNN
class SimpleCNN(nn.Module):
    def __init__(self, num_classes=10):
        super(SimpleCNN, self).__init__()
        self.features = nn.Sequential(
            nn.Conv2d(3, 32, kernel_size=3, padding=1),
            nn.BatchNorm2d(32),
            nn.ReLU(),
            nn.MaxPool2d(2, 2),  # 16x16
            nn.Conv2d(32, 64, kernel_size=3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.MaxPool2d(2, 2),  # 8x8
        )
        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Linear(64 * 8 * 8, 128),
            nn.ReLU(),
            nn.Dropout(0.5),
            nn.Linear(128, num_classes)
        )

    def forward(self, x):
        x = self.features(x)
        x = self.classifier(x)
        return x


# Define AlexNet
class AlexNet(nn.Module):
    """
    AlexNet variant for 32x32 inputs (SVHN / CIFAR style)
    """
    def __init__(self, num_classes=10):
        super(AlexNet, self).__init__()
        self.features = nn.Sequential(
            nn.Conv2d(3, 64, kernel_size=5, stride=1, padding=2),  
            nn.BatchNorm2d(64),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=2, stride=2),                  

            nn.Conv2d(64, 192, kernel_size=5, stride=1, padding=2), 
            nn.BatchNorm2d(192),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(kernel_size=2, stride=2),                  

            nn.Conv2d(192, 384, kernel_size=3, stride=1, padding=1),
            nn.ReLU(inplace=True),
            nn.Conv2d(384, 256, kernel_size=3, stride=1, padding=1),
            nn.ReLU(inplace=True),
            nn.Conv2d(256, 256, kernel_size=3, stride=1, padding=1),
            nn.ReLU(inplace=True),

            nn.MaxPool2d(kernel_size=2, stride=2)                   
        )
        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Linear(256 * 4 * 4, 1024),
            nn.ReLU(inplace=True),
            nn.Dropout(0.5),
            nn.Linear(1024, 512),
            nn.ReLU(inplace=True),
            nn.Dropout(0.5),
            nn.Linear(512, num_classes)
        )

    def forward(self, x):
        x = self.features(x)
        x = self.classifier(x)
        return x


# Training Utilities
def train_model(model, train_loader, test_loader, criterion, optimizer, epochs=5, device="cpu"):
    train_losses, val_losses, train_acc, val_acc = [], [], [], []
    model.to(device)

    for epoch in range(epochs):
        model.train()
        total_loss, correct, total = 0, 0, 0

        for imgs, labels in train_loader:
            imgs, labels = imgs.to(device), labels.to(device)
            optimizer.zero_grad()
            outputs = model(imgs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            total_loss += loss.item()
            _, preds = torch.max(outputs, 1)
            correct += (preds == labels).sum().item()
            total += labels.size(0)

        train_losses.append(total_loss / len(train_loader))
        train_acc.append(correct / total)

        # Validation
        model.eval()
        val_loss, val_correct, val_total = 0, 0, 0
        with torch.no_grad():
            for imgs, labels in test_loader:
                imgs, labels = imgs.to(device), labels.to(device)
                outputs = model(imgs)
                loss = criterion(outputs, labels)
                val_loss += loss.item()
                _, preds = torch.max(outputs, 1)
                val_correct += (preds == labels).sum().item()
                val_total += labels.size(0)
        val_losses.append(val_loss / len(test_loader))
        val_acc.append(val_correct / val_total)

        print(f"Epoch [{epoch+1}/{epochs}] "
              f"Train Loss: {train_losses[-1]:.4f}, Acc: {train_acc[-1]*100:.2f}% | "
              f"Val Loss: {val_losses[-1]:.4f}, Acc: {val_acc[-1]*100:.2f}%")

    return train_losses, val_losses, train_acc, val_acc

# Instantiate models and configs
device = "cuda" if torch.cuda.is_available() else "cpu"
print("Using device:", device)

criterion = nn.CrossEntropyLoss()

# Simple CNN
simple_model = SimpleCNN(num_classes=10)
optimizer_simple = optim.Adam(simple_model.parameters(), lr=0.001)

# AlexNet
alex_model = AlexNet(num_classes=10)
optimizer_alex = optim.Adam(alex_model.parameters(), lr=0.001)

# Train both models
start = time.time()
print("\nTraining SimpleCNN...")
simple_hist = train_model(simple_model, train_loader, test_loader, criterion, optimizer_simple, epochs=5, device=device)

print("\nTraining AlexNet...")
alex_hist = train_model(alex_model, train_loader, test_loader, criterion, optimizer_alex, epochs=5, device=device)
print("Total training time (min):", (time.time() - start)/60)

# Plot training curves
def plot_curves(hist, title):
    train_losses, val_losses, train_acc, val_acc = hist
    plt.figure(figsize=(10,4))
    plt.subplot(1,2,1)
    plt.plot(train_losses, label='Train Loss')
    plt.plot(val_losses, label='Val Loss')
    plt.title(f'{title} - Loss')
    plt.legend()
    plt.subplot(1,2,2)
    plt.plot(train_acc, label='Train Acc')
    plt.plot(val_acc, label='Val Acc')
    plt.title(f'{title} - Accuracy')
    plt.legend()
    plt.show()

plot_curves(simple_hist, "SimpleCNN")
plot_curves(alex_hist, "AlexNet")
```

**Question 5:** Using Keras/PyTorch, implement the Simple CNN and AlexNet to classify different didgits. Provide the details of the proposed architecture for both CNN and AlexNet.


```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report
import numpy as np
import torch
import time
import matplotlib.pyplot as plt

# Define evaluation function
def evaluate_model(model, data_loader, device="cpu"):
    model.eval()
    all_preds, all_labels = [], []
    with torch.no_grad():
        for imgs, labels in data_loader:
            imgs, labels = imgs.to(device), labels.to(device)
            outputs = model(imgs)
            _, preds = torch.max(outputs, 1)
            all_preds.extend(preds.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())
    all_preds, all_labels = np.array(all_preds), np.array(all_labels)
    acc = (all_preds == all_labels).mean()
    print("Accuracy:", acc)
    print(classification_report(all_labels, all_preds, digits=3, zero_division=0))
    return all_labels, all_preds, acc

# Evaluate both models
print("\nEvaluating SimpleCNN...")
y_true_cnn, y_pred_cnn, acc_cnn = evaluate_model(simple_model, test_loader, device)

print("\nEvaluating AlexNet...")
y_true_alex, y_pred_alex, acc_alex = evaluate_model(alex_model, test_loader, device)

# Confusion matrices
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
for ax, y_true, y_pred, title in zip(
    axes, 
    [y_true_cnn, y_true_alex], 
    [y_pred_cnn, y_pred_alex], 
    ["SimpleCNN Confusion Matrix", "AlexNet Confusion Matrix"]
):
    cm = confusion_matrix(y_true, y_pred)
    disp = ConfusionMatrixDisplay(cm)
    disp.plot(ax=ax, colorbar=False)
    ax.set_title(title)
plt.tight_layout()
plt.show()

# Compare model complexity
def count_params(model):
    return sum(p.numel() for p in model.parameters() if p.requires_grad)

cnn_params = count_params(simple_model)
alex_params = count_params(alex_model)

comparison = {
    "Model": ["SimpleCNN", "AlexNet"],
    "Parameters": [cnn_params, alex_params],
    "Test Accuracy": [acc_cnn, acc_alex],
}
import pandas as pd
pd.DataFrame(comparison)
```

Both models show very low accuracy because the dataset used for training is too small and only trained for five epochs.  
AlexNet has much more parameters but does not perform better because it needs more data and longer training to use its deep layers effectively.  
SimpleCNN, with fewer layers, trains faster and gives similar results.  
Overall, both models underfit the data, and increasing the training size and epochs would help AlexNet outperform SimpleCNN.

```python
# ==============================================================
# Q6: Train & Evaluate with Regularization + Optimization
# - Metrics: accuracy, precision, recall, f1, loss
# - Time per sample
# - Regularization: BatchNorm, Dropout (in models), Weight Decay
# - Optimization: Adam + CosineAnnealingLR, EarlyStopping, best model checkpoint
# ==============================================================

import time
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.metrics import classification_report, accuracy_score
import numpy as np
import copy

# -------- Config --------
EPOCHS = 20
BATCH_SIZE = 64  # already used in your DataLoader
WEIGHT_DECAY = 5e-4
LR = 1e-3
PATIENCE = 5               # early stopping
MODEL_NAME = "alexnet_q6.pt"  # or "simplecnn_q6.pt"
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"

# -------- Rebuild a fresh model to train for Q6 (choose one) --------
model_q6 = AlexNet(num_classes=10)   # or SimpleCNN(num_classes=10)
model_q6 = model_q6.to(DEVICE)

# Label-smoothing helps regularization on small data
criterion = nn.CrossEntropyLoss(label_smoothing=0.05)
optimizer = optim.Adam(model_q6.parameters(), lr=LR, weight_decay=WEIGHT_DECAY)
scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=EPOCHS)

def run_epoch(model, loader, train=True):
    model.train() if train else model.eval()
    total_loss, total, correct = 0.0, 0, 0
    all_preds, all_labels = [], []
    t0 = time.time()
    for imgs, labels in loader:
        imgs, labels = imgs.to(DEVICE), labels.to(DEVICE)
        if train:
            optimizer.zero_grad()

        outputs = model(imgs)
        loss = criterion(outputs, labels)

        if train:
            loss.backward()
            optimizer.step()

        total_loss += loss.item() * labels.size(0)
        _, preds = outputs.max(1)
        correct += (preds == labels).sum().item()
        total += labels.size(0)
        all_preds.append(preds.detach().cpu().numpy())
        all_labels.append(labels.detach().cpu().numpy())

    elapsed = time.time() - t0
    loss_avg = total_loss / total
    acc = correct / total
    y_pred = np.concatenate(all_preds)
    y_true = np.concatenate(all_labels)
    return loss_avg, acc, y_true, y_pred, elapsed, total  # elapsed for time per sample

# -------- Training with Early Stopping + Best Checkpoint --------
best_val_acc = 0.0
best_state = None
no_improve = 0

hist = {"train_loss": [], "val_loss": [], "train_acc": [], "val_acc": [], "time_per_sample": []}

for epoch in range(1, EPOCHS + 1):
    tr_loss, tr_acc, _, _, tr_time, tr_n = run_epoch(model_q6, train_loader, train=True)
    va_loss, va_acc, y_true, y_pred, va_time, va_n = run_epoch(model_q6, test_loader, train=False)

    scheduler.step()

    # time per sample (on training set)
    tps = tr_time / tr_n
    hist["train_loss"].append(tr_loss); hist["val_loss"].append(va_loss)
    hist["train_acc"].append(tr_acc);   hist["val_acc"].append(va_acc)
    hist["time_per_sample"].append(tps)

    print(f"Epoch {epoch:02d}/{EPOCHS} | "
          f"Train Loss {tr_loss:.4f} Acc {tr_acc*100:.2f}% | "
          f"Val Loss {va_loss:.4f} Acc {va_acc*100:.2f}% | "
          f"Time/sample {tps*1000:.2f} ms")

    # early stopping on val acc
    if va_acc > best_val_acc:
        best_val_acc = va_acc
        best_state = copy.deepcopy(model_q6.state_dict())
        torch.save(best_state, MODEL_NAME)
        no_improve = 0
    else:
        no_improve += 1
        if no_improve >= PATIENCE:
            print(f"Early stopping at epoch {epoch}. Best Val Acc = {best_val_acc*100:.2f}%")
            break

# -------- Load best checkpoint and final evaluation --------
model_q6.load_state_dict(torch.load(MODEL_NAME, map_location=DEVICE))
final_loss, final_acc, y_true, y_pred, _, _ = run_epoch(model_q6, test_loader, train=False)
print("\nFinal Evaluation on Test Set")
print(f"Loss: {final_loss:.4f} | Accuracy: {final_acc:.4f}")
print(classification_report(y_true, y_pred, digits=3, zero_division=0))

# Optional: plot curves
import matplotlib.pyplot as plt
plt.figure(figsize=(10,4))
plt.subplot(1,2,1); plt.plot(hist["train_loss"], label="Train"); plt.plot(hist["val_loss"], label="Val")
plt.title("Loss"); plt.legend()
plt.subplot(1,2,2); plt.plot(hist["train_acc"], label="Train"); plt.plot(hist["val_acc"], label="Val")
plt.title("Accuracy"); plt.legend(); plt.show()

print(f"Average training time per sample: {np.mean(hist['time_per_sample']):.6f} s")
```

The model shows stable training with early stopping and regularization (BatchNorm, Dropout, Weight Decay).  
Training time per sample is about 0.0044 s, and validation loss stays close to training loss, showing no overfitting.  
However, both training and validation accuracy remain low (â‰ˆ 29%), meaning the model still underfits due to the small dataset and short training.  
Regularization prevented overfitting, but further improvement needs more data, longer training, and learning-rate tuning.  
Overall, the model is well-optimized but limited by data scale rather than architecture or training stability.
